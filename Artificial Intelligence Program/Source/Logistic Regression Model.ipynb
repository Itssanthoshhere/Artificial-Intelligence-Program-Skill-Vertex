```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

###### Load dataset


```python
data = pd.read_csv("car_data.csv")  # Reads the CSV file into a DataFrame
data  # Display the first few rows of the DataFrame
```

###### Prepare feature matrix (X) and target vector (y)


```python
x = data.iloc[:, [2, 3]].values  # Selects columns Age and AnnualSalary as features
x  # Display the feature matrix
```

```python
y = data.iloc[:, 4].values  # Selects the Purchased column as the target variable
y # Display the target vector
```

###### Split the data into training and test sets

```python
from sklearn.model_selection import train_test_split
```

```python
'''
  1. x represents the feature matrix, y represents the target variable
  2. test_size specifies the proportion of data to include in the test set (30% in this case)
  3. random_state ensures reproducibility of the results by initializing the random number generator '''

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)
```

       `x_train` and `y_train` are used for training the model
       `x_test` and `y_test` are used for evaluating the model

###### Display the training feature matrix

```python
# The output shows the feature values used for training the model
print("x_train:")
x_train
```

###### Display the test feature matrix

```python
# The output shows the feature values used for testing the model
print("x_test:")
x_test 
```

###### Data Standardization with StandardScaler in Python


```python
from sklearn.preprocessing import StandardScaler

# Instantiate the StandardScaler
scaler = StandardScaler()

# Fit and transform the training data
x_train = scaler.fit_transform(x_train)

# Transform the test data
x_test = scaler.transform(x_test)
```

```python
# Output standardized training data
print("Standardized Training Data:\n", x_train)
```

```python
# Output standardized test data
print("Standardized Test Data:\n", x_test)
```

###### Training the Model:

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(x_train, y_train)
```

###### Making Predictions:

```python
y_pred = classifier.predict(x_test)
```

###### Viewing Predictions:

```python
y_pred
```

###### Confusion Matrix

  Computing the Confusion Matrix:

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix
cm
```

 Visualizing the Confusion Matrix:

```python
import seaborn as sns
cm = confusion_matrix(y_test, y_pred)
plt.title('Confusion Matrix Heatmap', fontsize=12)
sns.heatmap(cm, annot=True, fmt='d')
plt.show()
```
