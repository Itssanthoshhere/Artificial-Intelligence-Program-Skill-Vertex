#### 1. Import Libraries and Load Data:



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv("framingham.csv")
df
```

```python
# Display the first few rows of the dataset
df.head()
```

```python
# Display basic statistics of the dataset
df.describe()
```

```python
# Check for missing values
df.isna().sum()
```

#### 2. Data Preparation:



```python
# Define features (X) and target (y)
x = df.iloc[:, 0:15].values
y = df.iloc[:, 15].values
```

#### 3. Handle missing values


```python
# Use SimpleImputer to fill missing values with the mean
from sklearn.impute import SimpleImputer

# Handle missing values
imputer = SimpleImputer(strategy='mean')
x = imputer.fit_transform(x)
```

#### 4. Splitting Data:

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)
```

#### 5. Training KNN Classifier


```python
# Create and train the KNN classifier
from sklearn.ensemble import RandomForestClassifier

# Initialize and train the classifier
classifier = RandomForestClassifier(n_estimators=100, random_state=0)
classifier.fit(x_train, y_train)
```

#### 6. Making Predictions


```python
# Predict on the test set
y_pred = classifier.predict(x_test)
```

#### 7. Evaluating the Model


```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Accuracy score
print("Accuracy Score:", accuracy_score(y_test, y_pred))
```

```python
# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
conf_matrix_df = pd.DataFrame(data=cm, columns=['Predicted:0', 'Predicted:1'], index=['Actual:0', 'Actual:1'])

# Display confusion matrix
plt.figure(figsize=(10, 6))
sns.heatmap(conf_matrix_df, annot=True, fmt='d', cmap='Greens', linecolor='yellow', linewidths=1.5)
plt.title('Confusion Matrix')
plt.show()
```

```python
# Classification Report
print("Classification Report:\n", classification_report(y_test, y_pred))
```

#### 8. Visualizing Feature Importances

```python
# Extract feature importances
importances = classifier.feature_importances_
feature_names = df.columns[:-1]  # Exclude the target column
```

```python
# Create a DataFrame for feature importances
feature_importances_df = pd.DataFrame({
    'Feature': feature_names,
    'Importance': importances
}).sort_values(by='Importance', ascending=False)
```

```python
# Plot feature importances
plt.figure(figsize=(12, 8))
sns.barplot(x='Importance', y='Feature', data=feature_importances_df, palette='viridis')
plt.title('Feature Importances')
plt.show()
```
