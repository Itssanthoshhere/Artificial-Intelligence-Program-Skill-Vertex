### Creating a DataFrame

You can create a DataFrame from a dictionary of lists.

```python
# ğ—–ğ—¿ğ—²ğ—®ğ˜ğ—¶ğ—»ğ—´ ğ—® ğ——ğ—®ğ˜ğ—®ğ—™ğ—¿ğ—®ğ—ºğ—² ğ—³ğ—¿ğ—¼ğ—º ğ—® ğ—±ğ—¶ğ—°ğ˜ğ—¶ğ—¼ğ—»ğ—®ğ—¿ğ˜†
data = {'cars' : ['honda', 'hyundai', 'audi', 'bmw'],'range':[10,40,30,20]}

# ğ—–ğ—¿ğ—²ğ—®ğ˜ğ—¶ğ—»ğ—´ ğ˜ğ—µğ—² ğ——ğ—®ğ˜ğ—®ğ—™ğ—¿ğ—®ğ—ºğ—²
v = pd.DataFrame(data)

# ğ—£ğ—¿ğ—¶ğ—»ğ˜ğ—¶ğ—»ğ—´ ğ˜ğ—µğ—² ğ——ğ—®ğ˜ğ—®ğ—™ğ—¿ğ—®ğ—ºğ—²
print(v)
```

##### Explanation
1. ğ—–ğ—¿ğ—²ğ—®ğ˜ğ—¶ğ—»ğ—´ ğ—® ğ——ğ—®ğ˜ğ—®ğ—™ğ—¿ğ—®ğ—ºğ—²: The dictionary data contains two keys, cars and range, each associated with a list of values. The pd.DataFrame(data) function converts this dictionary into a DataFrame.
2. ğ—£ğ—¿ğ—¶ğ—»ğ˜ğ—¶ğ—»ğ—´ ğ˜ğ—µğ—² ğ——ğ—®ğ˜ğ—®ğ—™ğ—¿ğ—®ğ—ºğ—²: The print(v) statement displays the DataFrame.

### Working with Series?

A Series is a one-dimensional array-like object containing an array of data and an associated array of data labels, called its index.

```python
# ğ—–ğ—¿ğ—²ğ—®ğ˜ğ—¶ğ—»ğ—´ ğ—® ğ—¦ğ—²ğ—¿ğ—¶ğ—²ğ˜€ ğ—³ğ—¿ğ—¼ğ—º ğ—® ğ—¹ğ—¶ğ˜€ğ˜
a = [1,3,4,5]
var = pd.Series(a)

# ğ—£ğ—¿ğ—¶ğ—»ğ˜ğ—¶ğ—»ğ—´ ğ˜ğ—µğ—² ğ—¦ğ—²ğ—¿ğ—¶ğ—²ğ˜€
print(var)
```

##### Explanation
1. ğ—–ğ—¿ğ—²ğ—®ğ˜ğ—¶ğ—»ğ—´ ğ—® ğ—¦ğ—²ğ—¿ğ—¶ğ—²ğ˜€: The list a is converted into a Pandas Series using pd.Series(a).
2. ğ—£ğ—¿ğ—¶ğ—»ğ˜ğ—¶ğ—»ğ—´ ğ˜ğ—µğ—² ğ—¦ğ—²ğ—¿ğ—¶ğ—²ğ˜€: The print(var) statement displays the Series.

```python
# ğ—”ğ—°ğ—°ğ—²ğ˜€ğ˜€ğ—¶ğ—»ğ—´ ğ—®ğ—» ğ—²ğ—¹ğ—²ğ—ºğ—²ğ—»ğ˜ ğ—³ğ—¿ğ—¼ğ—º ğ˜ğ—µğ—² ğ—¦ğ—²ğ—¿ğ—¶ğ—²ğ˜€ ğ—¯ğ˜† ğ—¶ğ—»ğ—±ğ—²ğ˜…
print(var[1])
```

##### Explanation
1. ğ—”ğ—°ğ—°ğ—²ğ˜€ğ˜€ğ—¶ğ—»ğ—´ ğ—˜ğ—¹ğ—²ğ—ºğ—²ğ—»ğ˜ğ˜€: You can access elements in a Series using their index.

```python
# ğ—–ğ—¿ğ—²ğ—®ğ˜ğ—¶ğ—»ğ—´ ğ—® ğ—¦ğ—²ğ—¿ğ—¶ğ—²ğ˜€ ğ˜„ğ—¶ğ˜ğ—µ ğ—® ğ—°ğ˜‚ğ˜€ğ˜ğ—¼ğ—º ğ—¶ğ—»ğ—±ğ—²ğ˜…
b = pd.Series(a, index=['a', 'b', 'c', 'd'])

# ğ—£ğ—¿ğ—¶ğ—»ğ˜ğ—¶ğ—»ğ—´ ğ˜ğ—µğ—² ğ—¦ğ—²ğ—¿ğ—¶ğ—²ğ˜€ ğ˜„ğ—¶ğ˜ğ—µ ğ—® ğ—°ğ˜‚ğ˜€ğ˜ğ—¼ğ—º ğ—¶ğ—»ğ—±ğ—²ğ˜…
print(b)
```

##### Explanation
1. ğ—–ğ˜‚ğ˜€ğ˜ğ—¼ğ—º ğ—œğ—»ğ—±ğ—²ğ˜…: You can assign custom index labels to the elements of the Series.

```python
# ğ—–ğ—¿ğ—²ğ—®ğ˜ğ—¶ğ—»ğ—´ ğ—® ğ—¦ğ—²ğ—¿ğ—¶ğ—²ğ˜€ ğ—³ğ—¿ğ—¼ğ—º ğ—® ğ—±ğ—¶ğ—°ğ˜ğ—¶ğ—¼ğ—»ğ—®ğ—¿ğ˜†
stud = {'ram': 65, 'sita': 80, 'priya': 90}
vi = pd.Series(stud)

# ğ—£ğ—¿ğ—¶ğ—»ğ˜ğ—¶ğ—»ğ—´ ğ˜ğ—µğ—² ğ—¦ğ—²ğ—¿ğ—¶ğ—²ğ˜€
print(vi)
```

##### Explanation
1. ğ—¦ğ—²ğ—¿ğ—¶ğ—²ğ˜€ ğ—³ğ—¿ğ—¼ğ—º ğ——ğ—¶ğ—°ğ˜ğ—¶ğ—¼ğ—»ğ—®ğ—¿ğ˜†: A dictionary can be directly converted into a Series, where the keys become the index and the values become the data.

### Working with DataFrame

```python
import pandas as pd

# ğ—–ğ—¿ğ—²ğ—®ğ˜ğ—¶ğ—»ğ—´ ğ—® ğ——ğ—®ğ˜ğ—®ğ—™ğ—¿ğ—®ğ—ºğ—² ğ—³ğ—¿ğ—¼ğ—º ğ—® ğ—±ğ—¶ğ—°ğ˜ğ—¶ğ—¼ğ—»ğ—®ğ—¿ğ˜†
data = {'cars': ['honda', 'hyundai', 'audi', 'bmw'], 'range': [10, 40, 30, 20]}
v = pd.DataFrame(data)

# ğ—£ğ—¿ğ—¶ğ—»ğ˜ğ—¶ğ—»ğ—´ ğ˜ğ—µğ—² ğ——ğ—®ğ˜ğ—®ğ—™ğ—¿ğ—®ğ—ºğ—² ğ˜ğ—¼ ğ˜€ğ—²ğ—² ğ—¶ğ˜ğ˜€ ğ—°ğ—¼ğ—»ğ˜ğ—²ğ—»ğ˜
print(v)
```

```python
# ğ—¦ğ—²ğ—¹ğ—²ğ—°ğ˜ğ—¶ğ—»ğ—´ ğ—¿ğ—¼ğ˜„ğ˜€ ğ˜„ğ—¶ğ˜ğ—µ ğ—¶ğ—»ğ—±ğ—¶ğ—°ğ—²ğ˜€ ğŸ¬, ğŸ­, ğ—®ğ—»ğ—± ğŸ®
print(v.loc[[0, 1, 2]])
```

```python
# ğ—¦ğ—²ğ—¹ğ—²ğ—°ğ˜ğ—¶ğ—»ğ—´ ğ—¼ğ—»ğ—¹ğ˜† ğ˜ğ—µğ—² ğ—¿ğ—¼ğ˜„ ğ˜„ğ—¶ğ˜ğ—µ ğ—¶ğ—»ğ—±ğ—²ğ˜… ğŸ¬
print(v.loc[[0]])
```

### Loading Data from a CSV File

```python
import pandas as pd

# Reading the CSV file into a DataFrame
df = pd.read_csv('Mall_Customers.csv')

# Displaying the contents of the DataFrame
df
```

### Data Cleaning

#### Dropping Missing Values

```python
# Dropping rows with missing values
new_df = df.dropna()

# Printing the cleaned DataFrame
new_df
```

##### Explanation
1. ğ——ğ—¿ğ—¼ğ—½ğ—½ğ—¶ğ—»ğ—´ ğ—¥ğ—¼ğ˜„ğ˜€: The df.dropna() function removes rows with missing values.

```python
# Dropping rows with missing values in place
df.dropna(inplace=True)

# Printing the cleaned DataFrame
df
```

##### Explanation
1. ğ—œğ—»-ğ—£ğ—¹ğ—®ğ—°ğ—² ğ——ğ—¿ğ—¼ğ—½ğ—½ğ—¶ğ—»ğ—´: The inplace=True parameter modifies the original DataFrame directly.

#### Filling Missing Values

```python
# Filling missing values with a specified value
df.fillna(30, inplace=True)

# Printing the DataFrame with filled values
df
```

##### Explanation
1. ğ—™ğ—¶ğ—¹ğ—¹ğ—¶ğ—»ğ—´ ğ—©ğ—®ğ—¹ğ˜‚ğ—²ğ˜€: The df.fillna(value) function fills missing values with the specified value.

### Descriptive Statistics

#### Calculating Mean, Median, and ModeData Cleaning

```python
# Calculating the mean of a column
x = df['Age'].mean()
print(x)
```

```python
# Calculating the median of a column
y = df['Age'].median()
y
```

```python
# Calculating the mode of a column
z = df['Age'].mode()
z
```

##### Explanation
1. ğ— ğ—²ğ—®ğ—»: The mean() function calculates the average value.
2. ğ— ğ—²ğ—±ğ—¶ğ—®ğ—»: The median() function calculates the middle value.
3. ğ— ğ—¼ğ—±ğ—²: The mode() function calculates the most frequent value.
